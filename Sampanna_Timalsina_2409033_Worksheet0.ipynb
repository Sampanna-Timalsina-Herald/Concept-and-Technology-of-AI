```python
#RECURSION
#1
def sum_nested_list(nested_list):
"""
 Calculate the sum of all numbers in a nested list.
 This function takes a list that may contain integers and other nested lists.
 It recursively traverses the list and sums all the integers, no matter how deeply
 nested they are.
 Args:
 nested_list (list): A list that may contain integers or other lists of integers.
 Returns:
        int :The sum of integers of the list
"""
    total=0
    for element in nested_list:
        if isinstance(element,list):
            total += sum_nested_list(element)
        else:
            total += element
    return total
print(f"The sum of the integers of the nested list is:{sum_nested_list([1,[2,3,4],5])}")
```

```python
#2
def generate_permutations(s):
    """
    A function to compute all the unique permutations of a string.
    This function accepts a string and generates all its possible unique permutations.
    It also utilizes a helper function, `permute`, which accepts 4 parameters
    to generate the permutations.
    """
    def permute(result, s, end, start):
        if start == end:
            result.append("".join(s))
        else:
            seen = set()
            for i in range(start, end):
                if s[i] in seen:
                    continue
                seen.add(s[i])
                s[start], s[i] = s[i], s[start]
                permute(result, s, end, start + 1)
                s[start], s[i] = s[i], s[start]

    # If all characters are the same, return the input string as the only permutation
    if all(c == s[0] for c in s):
        return ["".join(s)]

    result = []
    permute(result, list(s), len(s), 0)
    return result

# Test the function
print(generate_permutations("ABC"))
```

```python
def calculate_directory_size(directory, exists=None):
    """
    A function to determine the size of a directory using recursion. 
    Arguments:
        directory: A dictionary with file names as keys and the space they occupy as values.
    Returns:
        int: The total size of the directory.
    """
    if exists is None:
        exists = set()
    
    total_space = 0
    
    for item, value in directory.items():
        if isinstance(value, dict):
            # Check if this subdirectory has already been visited
            if id(value) in exists:
                continue
            exists.add(id(value))
            # Recursive call for the subdirectory
            total_space += calculate_directory_size(value, exists)
        else:
            total_space += value
    
    return total_space

# Example directory structure
directory = {
    "file1.txt": 500,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 800,
        "file4.txt": 200
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 300
        },
        "file6.txt": 200
    }
}

# Calculate and print the size of the directory
print(f"The size of the directory is: {calculate_directory_size(directory)} KB")
```

```python
#Task 2 Dynamic Programming LongestCommonSubsequence(LCS):
def longest_common_subsequence(s1, s2):
"""
A function to find the length of the longest common subsequence. This function uses dynamic 
programming to compare the strings and determine the longest common subsequence.
Arguments:
    s1 (string): The first string to calculate the longest common subsequence.
    s2 (string): The second string to compare and find the longest common subsequence.
Returns:
    int: An integer representing the length of the longest common subsequence.
"""

    if len(s1)>len(s2):
        dp=[[0]*(len(s2)+1) for _ in range(len(s1)+1)]
    elif len(s1)<len(s2):
        dp=[[0]*(len(s1)+1) for _ in range(len(s2)+1)]
    for i in range(len(s1)-1,-1,-1):
       for j in range(len(s2)-1,-1,-1):
           if(s1[i] == s2[j]):
               dp[i][j]=1 + dp[i+1][j+1]
           else:
               dp[i][j]=max(dp[i][j+1],dp[i+1][j])
    return dp[0][0]
print(f"The longest common subsequence is:{longest_common_subsequence('abcde','ace')}")           
                
```

```python
#Task 3 Dynamic Programming Knapsack Problem:
def knapsack(weights,values,capacity):
"""
A function to determine the maximum value achievable within the given capacity.
Arguments:
    weights: A list of weights associated with the objects.
    values:  A list of values corresponding to the objects.
    capacity: An integer representing the maximum weight that can be stored.
Returns:
    int: The maximum value that can be achieved within the specified capacity.
"""
    dp=[[0]*(capacity+1) for _ in range(len(weights))]
    for i in range(len(weights)):
        for j in range(capacity+1):
            if weights[i] <= j:
                #storing the values according to the weights class and profit
                dp[i][j] = max(values[i]+dp[i-1][j-weights[i]],dp[i-1][j])
            else:
                dp[i][j] = dp[i-1][j]
    return(dp[len(weights)-1][capacity])
weights=[1,3,4,5]
values=[1,4,5,7]
capacity=7
print(f"The total value of items that can be carried:{knapsack(weights,values,capacity)}")
                
```
